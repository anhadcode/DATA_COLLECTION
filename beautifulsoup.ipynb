```python
import requests
```

```python
url= "https://leapscholar.com/uk/university-details/imperial-college-london/admissions-acceptance-rate"
r= requests.get(url)
from bs4 import BeautifulSoup
soup= BeautifulSoup(r.text, 'html.parser')
table= soup.table
data= {}

for row in soup.find_all("tr"):
    # Find all <td> elements in the row
    cols = row.find_all("td")
    if len(cols) == 2:  # Ensure there are two columns (key-value pair)
        key = cols[0].get_text(strip=True)  # Extract key (first column)
        value = cols[1].get_text(strip=True)  # Extract value (second column)
        data[key] = value  # Add to dictionary
print(data)
```

```python
import csv

# Save the dictionary to a CSV file
csv_file = "MIT_graduate.csv"
with open(csv_file, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Parameter", "Value"])  # Write the header
    for key, value in data.items():
        writer.writerow([key, value])  # Write each key-value pair

print(f"Data has been saved to {csv_file}")
```

```python
from bs4 import BeautifulSoup
import requests

# URL of the website to scrape
url = 'https://www.postgraduate.study.cam.ac.uk/apply/before/entry-requirements'

# Fetch the website content
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    html_content = response.text  # Get the HTML content

    # Parse the HTML content with Beautiful Soup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Extract all text from the HTML
    text = soup.get_text()

    # Print the extracted text
    print(text)
else:
    print(f"Failed to retrieve the website. Status code: {response.status_code}")
```
